```python
from PIL import Image
from IPython.display import display
import numpy as np

# Load the images
image1 = Image.open('img1.jpg')
image2 = Image.open('2nd.jpg')

# Resize images to the same dimensions if necessary
image1 = image1.resize((image2.width, image2.height))

# Display the images one by one
display(image1)
display(image2)

# Convert images to numpy arrays
image1_array = np.array(image1)
image2_array = np.array(image2)

# Perform logical operations
logical_and = np.bitwise_and(image1_array, image2_array)
logical_or = np.bitwise_or(image1_array, image2_array)
logical_xor = np.bitwise_xor(image1_array, image2_array)
logical_not1 = np.bitwise_not(image1_array)
logical_not2 = np.bitwise_not(image2_array)

# Display the results
display(Image.fromarray(logical_and))
display(Image.fromarray(logical_or))
display(Image.fromarray(logical_xor))
display(Image.fromarray(logical_not1))
display(Image.fromarray(logical_not2))

# Intersection between two images
intersection = np.bitwise_and(logical_not1, logical_not2)
intersection2 = np.bitwise_not(intersection)
display(Image.fromarray(intersection2))


```

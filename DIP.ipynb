```python
##GrayScale
from PIL import Image
import matplotlib.pyplot as plt
file = "Bird-2.png"
img = Image.open(file)

gray_img = img.convert("L")
gray_img.save("gray.jpeg")
```

```python
##Black n White
threshold=128
black_white_img=gray_img.point(lambda pixel: 0 if pixel < threshold else 255, '1')
black_white_img.save("./black_and_white.jpeg")
black_white_img.show()
```

```python
##Splitting to RGB Bands
from PIL import Image

# opening a multiband image (RGB specifically)
im = Image.open("Bird-2.png")


im1 = Image.Image.split(im)

# showing each band
red=im1[0].save("./red.jpg")
blue=im1[1].save("./blue.jpg")
green=im1[2].save("./green.jpg")
```

```python
##Image Profile
img = img.convert("L")
width, height = img.size
horizontal_profile = [sum(img.getpixel((x, y)) for y in range(height)) / height for x in range(width)]

# Plot the profile curve
plt.figure(figsize=(8, 4))
plt.plot(horizontal_profile, color='black')
plt.title('Horizontal Profile Curve')
plt.xlabel('Pixel Position')
plt.ylabel('Pixel Value')
plt.grid(True)
plt.show()
```

```python
img = img.convert("RGB")  # Convert to "RGB" mode if not already in "RGB"

# Split the image into red, green, and blue channels
red, green, blue = img.split()

print(img.mode)  # Output: RGB
print(red.mode)  # Output: L
print(green.mode)  # Output: L
print(blue.mode)  # Output: L

# Merge the channels in the correct order (original order of RGB)
new_image = Image.merge("RGB", (red, green, blue))
new_image.save('new_image_split.jpg')

print(new_image.mode)
```

```python
##Resize
new_size=(500, 300)
resized_img=img.resize(new_size)
resized_img.save("resize.jpeg")
resized_img.show()
```
